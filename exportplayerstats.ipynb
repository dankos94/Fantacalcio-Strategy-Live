```python
year=2024
midsizeName = "ITA.1"

dir1 = "/kaggle/input/espn-soccer-data"

# read in playerStats.csv
filename = dir1 + "/playerStats_data/playerStats_" + str(year) + "_" + midsizeName +".csv"
df_playerStats = pd.read_csv(filename)

# read in players.csv
filename = dir1 + "/base_data/players.csv"
df_players = pd.read_csv(filename)

# read in teams.csv
filename = dir1 + "/base_data/teams.csv"
df_teams = pd.read_csv(filename)

# read in leagues.csv
filename = dir1 + "/base_data/leagues.csv"
df_leagues = pd.read_csv(filename)
```

```python
df_playerStats = pd.merge(df_playerStats,df_players, how = "inner", left_on= ['athleteId'],right_on=['athleteId'])
df_playerStats = pd.merge(df_playerStats,df_teams, how = "inner", left_on= ['teamId'],right_on=['teamId'])
df_playerStats = pd.merge(df_playerStats,df_leagues, how = "inner", left_on= ['seasonType'],right_on=['seasonType'])
# print(df_playerStats.info())
df_playerStats_columns = ['seasonType','year_x','seasonName','league','name','displayName_x','dateOfBirth',
                          'displayWeight','displayHeight','jersey','citizenship','positionName',
                          'appearances_value','subIns_value','foulsCommitted_value','foulsSuffered_value',
                          'yellowCards_value','redCards_value','ownGoals_value','offsides_value',
                          'goalAssists_value','shotsOnTarget_value','totalShots_value','totalGoals_value',
                          'shotsFaced_value','saves_value','goalsConceded_value',
                          'timestamp_x']
df_playerStats = df_playerStats[df_playerStats_columns]
df_playerStats.rename(columns={
    'year_x':'year',
    'name':'team',
    'displayName_x':'playerName',
    'appearances_value':'appearances',
    'subIns_value':'subIns',
    'foulsCommitted_value':'foulsCommitted',
    'foulsSuffered_value':'foulsSuffered',
    'yellowCards_value':'yellowCards',
    'redCards_value':'redCards',
    'ownGoals_value':'ownGoals',
    'offsides_value':'offsides',
    'goalAssists_value':'goalAssists',
    'shotsOnTarget_value':'shotsOnTarget',
    'totalShots_value':'totalShots',
    'totalGoals_value':'totalGoals',
    'shotsFaced_value':'shotsFaced',
    'saves_value':'saves',
    'goalsConceded_value':'goalsConceded',
    'timestamp_x':'timestamp'
},inplace=True)
# print(df_playerStats.info())
# sort player stats by 'totalGoals' in descending order
df_playerStats = df_playerStats.sort_values('totalGoals',ascending=False)
df_playerStats
```

```python
df_playerStats.to_csv("out.csv", sep=';')
```
